### Cleaning Business Jargon

```python
!head ../Texts/business_jargon_raw.txt 
```

```python
business_jargon_list = []

with open("../Texts/business_jargon_raw.txt", "r") as f:
    for line in f:
        if line[:2] == "##" or len(line) == 1:
            continue
        elif len(line.split(" ")) > 4 or line[-2] == ".":
            continue
        else:
            business_jargon_list.append(line.lower().replace("\n",""))
            
print(business_jargon_list)
jargon_df = pd.DataFrame(
                        {'Jargon_Terms': business_jargon_list,
                         'Jargon_Type': "Business",
                        })
jargon_df.head()
```

```python
french_jargon_list = []

with open("../Texts/french_expressions.txt", "r") as f:
    for line in f:
        line = line.split("[")[0]
#         line = str(line.split("[")[0].encode('ASCII', 'ignore'))
        if line[:2] == "##" or len(line) == 1:
            continue
        elif len(line.split(" ")) > 4 or line[-2] == ".":
            continue
        else:
            french_jargon_list.append(line.lower().replace("\n",""))
            
print(french_jargon_list)
tmp_jargon_df = pd.DataFrame(
                        {'Jargon_Terms': french_jargon_list,
                         'Jargon_Type': "French",
                        })
jargon_df = jargon_df.append(tmp_jargon_df)
jargon_df
```

### Legal Jargon

```python
!head ../Texts/legal_jargon_raw.txt 
```

```python
legal_jargon_list = []

with open("../Texts/legal_jargon_raw.txt", "r") as f:
    for line in f:
        line = line.split(":")[0]
#         print(line)
#         line = str(line.split("[")[0].encode('ASCII', 'ignore'))
        if line[:2] == "##" or len(line) <= 2 or "|" in line:
            continue
#         elif len(line.split(" ")) > 4 or line[-2] == ".":
#             continue
        else:
            legal_jargon_list.append(line.lower().replace("\n",""))
            
# print(legal_jargon_list)
tmp_jargon_df = pd.DataFrame(
                        {'Jargon_Terms': legal_jargon_list,
                         'Jargon_Type': "Legal",
                        })
jargon_df = jargon_df.append(tmp_jargon_df)
jargon_df
```
